# 1. Importing Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2. Reading the csv file

```python
df = pd.read_csv(r"Airline_customer_satisfaction.csv")
df = df.sample(300)
```

# 4. Data cleaning

```python
# removing unwanted features
df.drop(["Departure/Arrival time convenient", 
         "Gate location", 
         "Online support", 
         "Ease of Online booking", 
         "On-board service", 
         "Baggage handling", 
         "Checkin service", 
         "Online boarding", 
         "Arrival Delay in Minutes"], 
        axis=1, inplace=True)
```

# Flight distance and satisfaction analysis

```python
filtered_df = df[["satisfaction","Flight Distance"]].set_index("satisfaction")
```

```python
filtered_df
```

```python
df["Flight Distance"].dtype
```

```python
df["Flight Distance"].value_counts()
```

```python
sns.swarmplot(data=df, y="Flight Distance", x="satisfaction", hue="satisfaction")
```

```python
plot1 = sns.histplot(
    data = df,
    x="Flight Distance", 
    hue="satisfaction",
    multiple="fill",
    palette=["AliceBlue", "Brown"],
    binwidth=1000,
)

plt.title("Proportion of Satisfied Customers by Flight Distance")
plot1.set_ylabel("Proportion")
plt.savefig("Flight Distance and Satisfaction distribution.jpg", format="jpg", dpi = 275)
```
